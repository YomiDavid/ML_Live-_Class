```python
X = df.drop('Class', axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=101)
```

```python
 from sklearn.model_selection import GridSearchCV
 from sklearn.ensemble import RandomForestClassifier
```

```python

n_estimators = [64,100,125,200]
max_features = [2,3,4]
oob_score = [True, False]
bootstrap = [True, False]

param_grid = {'n_estimators':n_estimators, 'max_features':max_features,'oob_score':oob_score,'bootstrap':bootstrap}
```

```python
forest_model = RandomForestClassifier()
```

```python
grid_model = GridSearchCV(forest_model,param_grid=param_grid,verbose=2)
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_params_
```

```python
y_pred = grid_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
plot_confusion_matrix(grid_model, X_test, y_test)
```

```python
n_miss = np.sum(y_pred != y_test)
n_miss
```

```python
from sklearn.metrics import accuracy_score
errors = []
misclassification = []

for n in range(1,200):
    rfc = RandomForestClassifier(n_estimators=n,max_features=2)
    rfc.fit(X_train,y_train)
    preds = rfc.predict(X_test)
    error = 1 - accuracy_score(y_test,preds)
    n_missed = np.sum(preds != y_test)

    errors.append(error)
    misclassification.append(n_missed)
```

```python
plt.plot(range(1,200),errors)
```

```python
plt.plot(range(1,200),misclassification)
```
