# CATEGORICAL ENCODING

```python
X1 = df.drop('species', axis =1)
```

```python
y = df['species']
```

```python
X = pd.get_dummies(X1,drop_first=True)
X.head()
```

```python
X.shape
```

# MODEL BUILDING

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report, plot_confusion_matrix, roc_curve
```

```python
print(classification_report(y_test,y_pred))
```

```python
plot_confusion_matrix(model,X_test,y_test)
```

```python
model.feature_importances_
```

```python
X.columns
```

```python
pd.DataFrame(index=X.columns, data=model.feature_importances_, columns=['Feature Importance']).sort_values('Feature Importance')
```

create a function to explore the hyperparameter to see the tree structure and classification report for multiple models

```python
def report_model(model):
    print(classification_report(y_test, y_pred))
    print('\n')
    plt.figure(figsize=(8,10), dpi=150)
    plot_tree(model, feature_names=X.columns, filled =True);
```

```python
pruned_tree = DecisionTreeClassifier(max_depth=3 )
```

```python
pruned_tree.fit(X_train, y_train)
```

```python
y_pred= pruned_tree.predict(X_test)
```

```python
report_model(pruned_tree)
```

```python
# to check the max number of leaf nodes
max_leaf_tree = DecisionTreeClassifier(max_leaf_nodes=3)
max_leaf_tree.fit(X_train,y_train)
y_pred = max_leaf_tree.predict(X_test)
```

```python
report_model(max_leaf_tree)
```

```python
entropy_tree = DecisionTreeClassifier(criterion='entropy')
```

```python
entropy_tree.fit(X_train,y_train)
y_pred = entropy_tree.predict(X_test)
```

```python
report_model(entropy_tree)
```
